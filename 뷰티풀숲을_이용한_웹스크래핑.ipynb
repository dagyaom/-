```python
# 요청 보내기, 응답받기.

import requests   # 웹과 소통하는 라이브러리 불러오기(파이썬에서 HTTP 요청 보내기)
response = requests.get('https://movie.naver.com/movie/bi/mi/basic.naver?code=202673')  #구글에 요청보내기'  #메소드 get : 뭔가를 가지고오는 메서드. 
print(response)  # 정상적으로 연결되면 HTTP 응답 객체(Response 타입)가 리턴된다.  : <Response [200]>
print(type(response))   # <class 'requests.models.Response'> : requests 라이브러리의 Response타입임을 알 수 있다.

# 응답 성공 여부
print(response.status_code)  # 응답의 상태코드 확인가능. https://developer.mozilla.org/ko/docs/Web/HTTP/Status
```

```python
#  응답 성공 여부 : requests 라이브러리에 있는 HTTPError 를 통해서 에러가 발생했을 때 HTTP 에러만 따로 빼주는 작업

import requests
from requests.exceptions import HTTPError

url = 'https://movie.naver.com/movie/bi/mi/basic.naver?code=202673'

try :
    resp = requests.get(url)

    resp.raise_for_status()  # ok 코드가 아닐 경우 에러발동

except HTTPError:
    print('HTTP 에러가 발생했습니다.')

except Exception as Err:
    print('다른 에러가 발생했습니다.')

else :
    print('성공')

# try, except, else : https://gomguard.tistory.com/122
```

```python
# 응답내용 : 하나의 HTML 파일. 하나의 문자열로 되어있음. 

resp.text
```

## 돌려받은 응답 내용을 파싱하고 정보얻기 : BeautifulSoup 라이브러리 사용

```python
# 설치하기  # 코랩에서는 뷰티풀숩이 안된다.. https://stackoverflow.com/questions/5663980/importerror-no-module-named-beautifulsoup 이걸로 해결.
!pip install bs4
```

```python
# 기본파싱
# 파싱 : 파싱 (parsing) 이란 문자열로 구성된 특정 문서들 (HTML, XML) 등을 파이썬에서도 쉽게 사용할 수 있도록 변환해주는 작업
# 기본적으로 사용할 수 있는 파서는 'html.parser'

import requests
from bs4 import BeautifulSoup

url = 'https://movie.naver.com/movie/bi/mi/basic.naver?code=202673'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')
soup  #따옴표가 없어짐.
```

```python
# 요소 찾기 : find 및 find_all

wrap_element = soup.find(id='wrap')
# id 는 주로 한번만 사용이 되기 때문에 find 를 사용합니다.
wrap_element
```

```python
basic_elements = soup.find_all(class_='basic')
basic_elements
# class 를 이용해 찾을 때에는 class 가 아닌 뒤에 밑줄 '_' 을 추가해야 합니다. 추가하지 않게 되는 경우에는 파이썬의 class 로 인식하게 되기 때문에 구별해줘야 합니다.
```

```python
basic_elements = soup.find_all(class_='basic')
for basic_el in basic_elements:
    print(basic_el.find(class_='score_result'))
```

```python
# 태그 활용
basic_elements = soup.find_all('div', class_='basic')
basic_elements
```

```python
# string 활용
soup.find_all(string = '_sympathyButton')
```

```python

soup.find_all(string = lambda text: 'raining' in text.lower())  # lambda인자인 text를 소문자로 바꾼 것 중에서 raining이라는 문자열이 있다.


# string 을 사용하면 원하는 정보를 찾을 수는 있지만 각 요소의 .string 속성을 불러오는 것이기 때문에 요소가 아닌 문자열로 리턴이 됩니다. 
# 따라서 하나의 요소로 받기 위해서는 태그도 같이 추가해줘야 합니다.

soup.find_all('h3', string='raining')


# lambda : https://blockdmask.tistory.com/520
#in, not in 연산자 : https://pydole.tistory.com/entry/Python-%ED%8F%AC%ED%95%A8Containment-%EC%97%B0%EC%82%B0%EC%9E%90-in-not-in
```

```python
# HTML 에서 'p' 태그 내부 글을 얻으려면 text 속성을 사용할 수 있습니다.

# <p class='cat'>This is a p-cat</p>
```

```python
cat_el = soup.find('p', class_='cat')

cat_el.text #=> 'This is a p-cat'
```

```python
# 불필요한 띄어쓰기가 있을 수 있습니다. 그럴 때에는 파이썬의 strip 메소드를 사용해서 정리해줄 수 있습니다.

cat_el.text.strip()
```
