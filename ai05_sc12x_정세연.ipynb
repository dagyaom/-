### 3) 아래의 특성들을 만드세요

1. **TotalMntProducts** - 최근 2년간 구매한 모든 제품 가격들을 합한 값을 가진 계산한 특성입니다

2. **TotalNumPurchases** - 구매횟수를 모두 합친 값을 가진 특성입니다

3. **DaysSinceJoined** - 오늘 날짜를 기준으로 가입한 기간을 보여주는 특성입니다 

4. **VIP** - 누적 구매 금액이 $2,000 이상을 넘고, 오늘을 기준으로 가입기간이 5년 이상이 된다면 1, 아니면 0의 값을 가진 칼럼을 만듭니다

```python
# 1번 답
df['TotalMntProducts'] = df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds']
df.head()

# groupby 가 무엇이였나.
# 특정 행:행 잘라서 합계하는 법.
```

```python
# 2번 답
df['TotalNumPurchases'] = df['NumDealsPurchases']+df['NumWebPurchases']+df['NumStorePurchases']
df.head()
```

```python
# 3번 답
import datetime as dt
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df['DaysSinceJoined'] = pd.Timestamp.today() - df['Dt_Customer']
# df['DaysSinceJoined'] = dt.datetime.now() - df['Dt_Customer']
df.head()

# 시계열 데이터 다루기 : https://datascienceschool.net/01%20python/04.08%20%EC%8B%9C%EA%B3%84%EC%97%B4%20%EC%9E%90%EB%A3%8C%20%EB%8B%A4%EB%A3%A8%EA%B8%B0.html
#현재시간 : https://rfriend.tistory.com/497
 #판다스 날짜열로 바꾸기. 파이썬 날짜차이 계산으로 datatime 키워드 알게됨 --> 공식문서 참조해서 거꾸로 된 날짜 변환방법 알게 됨.
 # 현재날짜나오는 명령어, 문자열 시계로 바꾸기 까지는 알았는데, 두가지를 어떻게 조합해야 할지를 몰라서 못푼 문제.
```

```python
df.dtypes
```

```python
# 4번 답 :누적 구매 금액이 $2,000 이상을 넘고, 오늘을 기준으로 가입기간이 5년 이상이 된다면 1, 아니면 0의 값을 가진 칼럼
# ★ 이부분은 잘 모르겠다. 

df.loc[((df['TotalMntProducts']>=2000)&(df['DaysSinceJoined']>=pd.Timedelta('1825 days'))), 'VIP'] = 1
df.loc[~((df['TotalMntProducts']>=2000)&(df['DaysSinceJoined']>=pd.Timedelta('1825 days'))), 'VIP'] = 0
df
```
